```python
import pandas as pd
```

```python
df = pd.read_csv("train.csv")
```

```python
df
```

```python
df.drop(columns=["Id", "Utilities", "Alley", "MasVnrType", "FireplaceQu", "Fence", "MiscFeature", "Neighborhood", "Condition2", "YearBuilt", "YearRemodAdd", "RoofMatl", "Heating", "CentralAir", "LowQualFinSF", "PoolArea", "YrSold", "PoolQC"], inplace=True)
```

```python
df.info()
```

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
import numpy as np
```

```python
y = df["SalePrice"]
X = df.drop(columns=["SalePrice"])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)
```

```python
X_train_num = X_train.drop(columns=[
    'MSZoning', 'Street', 'LotShape', 'LandContour', 'LotConfig', 'LandSlope',
    'Condition1', 'BldgType', 'HouseStyle', 'RoofStyle', 'Exterior1st', 'Exterior2nd',
    'ExterQual', 'ExterCond', 'Foundation', 'BsmtQual', 'BsmtCond',
    'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2', 'HeatingQC', 'Electrical',
    'KitchenQual', 'Functional', 'GarageType', 'GarageFinish',
    'GarageQual', 'GarageCond', 'SaleType', 'SaleCondition', 'PavedDrive'
])
X_train_categ = X_train[[
    'MSZoning', 'Street', 'LotShape', 'LandContour', 'LotConfig', 'LandSlope',
    'Condition1', 'BldgType', 'HouseStyle', 'RoofStyle', 'Exterior1st', 'Exterior2nd',
    'ExterQual', 'ExterCond', 'Foundation', 'BsmtQual', 'BsmtCond',
    'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2', 'HeatingQC', 'Electrical',
    'KitchenQual', 'Functional', 'GarageType', 'GarageFinish',
    'GarageQual', 'GarageCond', 'SaleType', 'SaleCondition', 'PavedDrive'
]]
X_test_num = X_test.drop(columns=[
    'MSZoning', 'Street', 'LotShape', 'LandContour', 'LotConfig', 'LandSlope',
    'Condition1', 'BldgType', 'HouseStyle', 'RoofStyle', 'Exterior1st', 'Exterior2nd',
    'ExterQual', 'ExterCond', 'Foundation', 'BsmtQual', 'BsmtCond',
    'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2', 'HeatingQC', 'Electrical',
    'KitchenQual', 'Functional', 'GarageType', 'GarageFinish',
    'GarageQual', 'GarageCond', 'SaleType', 'SaleCondition', 'PavedDrive'
])
X_test_categ = X_test[[
    'MSZoning', 'Street', 'LotShape', 'LandContour', 'LotConfig', 'LandSlope',
    'Condition1', 'BldgType', 'HouseStyle', 'RoofStyle', 'Exterior1st', 'Exterior2nd',
    'ExterQual', 'ExterCond', 'Foundation', 'BsmtQual', 'BsmtCond',
    'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2', 'HeatingQC', 'Electrical',
    'KitchenQual', 'Functional', 'GarageType', 'GarageFinish',
    'GarageQual', 'GarageCond', 'SaleType', 'SaleCondition', 'PavedDrive'
]]
```

```python
imp = IterativeImputer(max_iter=10, random_state=123, sample_posterior=True)
X_train_num_transformed = imp.fit_transform(X_train_num)
X_test_num_transformed = imp.transform(X_test_num)
```

```python
X_train_categ.fillna(df.mode().iloc[0], inplace=True)
X_test_categ.fillna(df.mode().iloc[0], inplace=True)
```

```python
X_train_categ.info()
```

```python
X_train_num_transformed_df = pd.DataFrame(X_train_num_transformed, columns=X_train_num.columns)
X_test_num_transformed_df = pd.DataFrame(X_test_num_transformed, columns=X_test_num.columns)
```

```python
X_train_categ.index = range(1168)
X_test_categ.index = range(292)
```

```python
X_train_imputed = pd.concat([X_train_categ, X_train_num_transformed_df], axis=1)
X_test_imputed = pd.concat([X_test_categ, X_test_num_transformed_df], axis=1)
```

```python
l = [
    'MSZoning', 'Street', 'LotShape', 'LandContour', 'LotConfig', 'LandSlope',
    'Condition1', 'BldgType', 'HouseStyle', 'RoofStyle', 'Exterior1st', 'Exterior2nd',
    'ExterQual', 'ExterCond', 'Foundation', 'BsmtQual', 'BsmtCond',
    'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2', 'HeatingQC', 'Electrical',
    'KitchenQual', 'Functional', 'GarageType', 'GarageFinish',
    'GarageQual', 'GarageCond', 'SaleType', 'SaleCondition', 'PavedDrive'
]
```

```python
for col in X_train_imputed.columns:
    if col in l:
        ohe1 = OneHotEncoder(sparse=False)
        encoded_msz = ohe1.fit_transform(X_train_imputed[[col]])
        encoded_msz = pd.DataFrame(encoded_msz, columns=ohe1.get_feature_names_out([col]))
        X_train_imputed.drop(columns=[col], inplace=True)
        X_train_imputed = pd.concat([X_train_imputed, encoded_msz.iloc[:, 1:]], axis=1)
        encoded_t = ohe1.fit_transform(X_test_imputed[[col]])
        encoded_t = pd.DataFrame(encoded_t, columns=ohe1.get_feature_names_out([col]))
        X_test_imputed.drop(columns=[col], inplace=True)
        X_test_imputed = pd.concat([X_test_imputed, encoded_t.iloc[:, 1:]], axis=1)
```
